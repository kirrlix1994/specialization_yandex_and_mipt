## Вопрос 2. 
Часто в качестве последнего слоя в задаче классификации используется слой softmax который выглядит как:

$h_j=e^z_j / ∑e^z_i$ , где $z_i$ входы этого слоя, а $h_j$ выходы. Номер, выхода с максимальным значением соответствует номеру предсказанного класса.

Какие из утверждений верны:

* выходы softmax слоя всегда ≥0
* выходы softmax слоя всегда ≤1
* softmax монотонная функция, то есть выход с максимальным значением соответствует максимальному значению входа
* сумма выходов слоя равна 1
