## Вопрос 7. 
Для повышения контраста можно использовать разнообразные функции преобразования яркости. Такие функции $g(x)$ должны удовлетворять единственному условию - их выход должен лежать в диапазоне [0, 255], для любого $x$ из диапазона [0,255].

Если в качестве такой функции выбрать семейство $c log(1 + a*x)$, где $x$ - яркость входного изображения, а $a$ и $c$ константы.

То такие функции:

* увеличивают контраст изображения
* уменьшают контраст изображения
* увеличивают в темных областях и уменьшают в светлых
* уменьшают в темных областях и увеличивают в светлых
